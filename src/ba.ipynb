```python
# password checker -

def validate_password(password):
    """Validates password meets complexity requirements
        Returns true if valid, raises ValueError if not."""
    
    has_upper = False
    has_lower = False
    has_number = False
    has_special = False

    if len(password) < 6:
        raise ValueError("Password must be at least 6 characters long.")

    for char in password:
        if char.isupper():
            has_upper = True

        if char.islower():
            has_lower = True          

        if char.isnumeric():
            has_number = True 
        
        if not char.isalnum():
            has_special = True

    if not has_upper:
        raise ValueError("Password must contain at least one uppercase letter.")
    
    if not has_special:
        raise ValueError("Password must contain at least one special character (!@#$%^&*, etc.)")

    if not has_number:    
        raise ValueError("Password must contain at least one number.")
    
    if not has_lower:
        raise ValueError("Password must contain at least one lowercase letter.")

    return True

validate_password("Timothy@123!")
```

```python
# finding the highest sequential id -
students = ["STU001", "STU005", "STU003", "STU009", "STU002"]
orders = ["ORD-0005", "ORD-0012", "ORD-0003", "ORD-0020", "ORD-0008"]


def next_student(lst):
    latest = []

    for id in lst:
        prefix = id[:3]
        number_str = id[3:]
        number_int = int(number_str)
        latest.append(number_int)

    maximum = max(latest)
    next_str = str(maximum + 1).zfill(3)
    inv_next = prefix + next_str
    return inv_next


def next_order(lst):
    empty_list = []

    for order in lst:
        prefix = order[:4]
        number = int(order[4:])
        empty_list.append(number)
    
    latest_order = max(empty_list)
    next_ordr = latest_order + 1
    next = prefix + str(next_ordr).zfill(4)
    return next

print(next_student(students))
print(next_order(orders))
```

```python
# user authentication system -
#1. User doesn't have an account -> Option to make a new account. Saves account in database.
#2. User has an account -> Option to login into account. Make changes. Save updates.

accounts = load_accounts()  # this loads client's database into memory.

# user tries to login with username and password.

def authenticate_user(username, password, accounts):
    """Checks if a username exists in file and authenticates access if password matches.

    Args:
        username (str): This is user's desired login id/username
        password (str): This is their desired password
        accounts (_type_): 
    
    Returns: Bank
        If the username and password matches with user's information then it returns the BankAccount object. Else, it returns None.
    """
    if username in accounts:
        user_account = accounts[username]           # selects the object key (username)
        if user_account.password == password:       # checks if the password matches.
            return user_account                     # grants access to user's account.
        else:
            return None
    else:
        return None
authenticate_user("grufru", "halleluah", accounts)
```

```python
# create new user account -
# constants:
NEW_ACCOUNT_BALANCE = 0.0
def create_new_account(username: str, password: str, name: str, accounts: dict) -> bool:
    """Creates a new bank account for the new client.

    Args:
        username (str): desired unique username.
        password (str): password that must certain requirements.
        name (str): full name of account holder.
        accounts (dict): Dictionary of existing BankAccount objects.

    Returns:
        bool: True if account created successfully, False otherwise.
    """
    # checking if username exists in the dictionary already.
    if username in accounts:
        return False
    
    # if it doesn't exist, validating if password meets requirements.
    try:
        validate_password(password)  
    except ValueError: 
        return False
    
    account_number = generate_account_number(accounts)
    new_account = BankAccount(username, password, name, account_number, NEW_ACCOUNT_BALANCE)

    # add new account to dictionary
    accounts[username] = new_account

    # save new account to database
    save_accounts(accounts)

    return True

```
